## Segment

```python
# Paths
path2load='/Volumes/Sandisk_SD/Work/IZFP/Laser/2023_iHub/02_QASS'
path2save='/Volumes/Sandisk_SD/Work/IZFP/Laser/2023_iHub/05_Evaluation/test'
```

```python
# Delete the files first
delete = False

if delete == True:
    for chNo in [1, 2, 3]:
        delete_all_files(f'{path2save}/ch{chNo}')
        print(f'Delete files in {path2save}/ch{chNo}')
        print(os.listdir(f'{path2save}/ch{chNo}'))
```

```python
# Instantiate the segmentor
fs = 6.25* 10**6 # sampling frequency [Hz]

segmenter = Segmenter(metadf)
segmenter.set_path(path2load=path2load, path2save=path2save)
segmenter.set_segmentation_params(fs=fs, past=0.0, future=0.0)
```

```python
for trialNo in metadf['trial'].unique()[:3]:
    segmenter.trialNo = trialNo
    segmenter.meta_data_handling()
    
    ch1 = segmenter.load(chNo=1, clip=False, verbose=True)
    segmenter.segment_and_save(ch1, verbose=False)
    ch2 = segmenter.load(chNo=2, clip=False, verbose=True)
    segmenter.segment_and_save(ch2, verbose=False)
    ch3 = segmenter.load(chNo=3, clip=False, verbose=True)
    segmenter.segment_and_save(ch3, verbose=False)
    
    print('+==========================================+')
```
